<font color="" face="Times New Roman" size="5px">**Definición 1.** *Un proceso estocástico es una colección de variables aleatorias $\{X_t: t \in T\}$ parametrizada por un conjunto $T$, llamado espacio parametral, en donde las variables toman valores en un conjunto $S$ llamado espacio de estados.*

<font color="" face="Comic Sans MS">

Considérese un sistema que puede cambiar de un estado a otro, a través del tiempo gracias a una ley de movimiento. También considere que el sistema evoluciona de una manera no determinista, sino, azarosa, entonces obtenemos una colección de variables aleatorias que dependen de un parámetro $t$ que nos ayuda a indicar el estado $s$ en donde se encuentra nuestra variable.

La definición de proceso estocástico toma como base un espacio de probabilidad $(\Omega, \mathscr{F}, \mathbb P) $ Y es importante tenerlo en cuenta, para enunciar de una manera más específica los resultados que se presentan más adelante.

Particularmente, un proceso estocástico puede clasificarse según dos tipos: Procesos a tiempo discreto y Procesos a tiempo continuo. Es importante señalar la diferencia entre el espacio de estados de una variable aleatoria y el espacio parametral. El espacio de estados también se puede pensar como un soporte de nuestra sucesión de variables parametrizada por el espacio $T$. Es decir, el proceso puede tener un soporte finito o no. En cambio, el espacio parametral discreto, tiene sus valores en los naturales o en los enteros $\mathbb{( N, Z)}$; en cambio, el espacio continuo generalmente se indica sólo bajo el conjunto: $\{t\in T| t\geq0\}$.

<font color="" face="Comic Sans MS">

**Procesos de Markov**

Estos tipos de procesos son modelos de mucha importancia y son los de mayor estudio en la rama de la probabilidad. Suponiendo que se conoce el estado presente del sistema, entonces los estados anteriores no tienen influencia en los estados futuros del sistema. Esta condición se llama propiedad de Markov y puede expresarse de la siguiente forma: Para cualesquiera tiempos $0\leq s\leq t$ y sea $A\in B(\mathbb R)$:  

$$P( X_t \in A| \mathcal F_s) = P(X_t \in A | X_s) $$

<font color="" face="Comic Sans MS">

Para poder hablar de variables aleatorias con valores en el espacio de estados $S$, es necesario asociar a este conjunto una $\sigma$-álgebra.
Considerando que $S$ es un subconjunto de $\mathbb R$, puede tomarse a la $\sigma$-álgebra de Borel de R restringida a S, es decir, $S \cap \mathscr B(\mathbb R)$.

Un proceso estocástico puede considerarse como una función de dos variables: $$ X: T  x  \Omega → S $$
Así, al par ordenado $(t,\omega)$ se le asocia el valor $X(t,\omega)$, simplificando la notación se escribe: $ X_t(\omega) $. Para cada valor de $t$ en $T$, el mapeo $\omega → X_t(\omega)$ es una v.a. Mientras que para cada $\omega$ en $\Omega$ fijo, la función $t \rightarrow X_t(\omega)$ es llamada una trayectoria o realización del proceso. Es por ello que a veces se define un proceso estocástico como una función aleatoria. Una de tales trayectorias típicas que, además cuenta con la propiedad de ser continua se muestra en la figura siguiente y corresponde a una trayectoria de un Movimiento Browniano.

<font color="" face="Comic Sans MS">

**Movimiento Browniano**

**Definición 3.** Un movimiento Browniano estándar unidimensional es un proceso estocástico $\{B_t: t\geq 0\}$ tal que:


1.   $B_0 = 0$ casi seguramente.
2.   Las trayectorias $t → B_t$ son continuas.
3.   El proceso tiene incrementos independientes.
4.   La variable $B_t - B_s$ tiene distribución $N(0,t-s)$ para $0\leq s\leq t$.

Las condiciones que aparecen en esta definición fueron dadas con consecuencia directa de las observaciones del fenómeno físico que explicó Einstein. Sin embargo, no garantiza que tal objeto matemático exista. En 1923 el matemático norteamericano Norbert Wiener demostró la existencia de un proceso con tales condiciones. Por lo que también es llamado así, Proceso De Wiener.

Ahora veremos algunos resultados importantes de este proceso.

Función de Probabilidad de transición:

En particular, la probabilidad de que un movimiento Browniano que inicia en x se encuentre en un conjunto $A \subseteq \mathbb R$ después de t unidades de tiempo es: $$ p(t,x,A) = \int_A p(t,x,y)dy  $$

Donde también definimos: $$p(t,x,y) = \frac{1}{\sqrt{2\pi\sigma ^2t }} e^{-(y-x)^2 / 2 \sigma ^2t}$$


De la propiedad 4, podemos hacer $s=0$ y entonces obtenemos que en particular la variable $B_t \backsim N(0,t)$. y por tanto:
1.  $E(B_t) = 0$
2.  $Var(B_t) = E(B_t^2) = t$
3.  $Var(B_t-B_s) = E((B_t-B_s)^2) = t-s$

También es importante resaltar que el Proceso de Wiener cumple con la propiedad de Markov.

Una ecuación diferencial ordinaria es del estilo: $$a(x,t) = \frac{dx}{dt}  $$

Se podría pensar a esta ecuación como una forma degenerada de una ecuación diferencial estocástica en la ausencia de aleatoriedad. Vayamos a esto. Es útil recordar unas de sus propiedades básicas. Podemos reescribirlo en los símbolos de las diferenciales así:
 $$dx = a(t,x)dt $$
 o de manera más exacta como un ecuación integral:
 $$x(t) = x_0 + \int_{t_0}^t a(s,x(s))ds $$
 Donde $x(t) = x(t;x_0,t_0)$ es una soluxión que satisface la condición $x(t_0) = x_0$. Estas soluciones están entonces relacionadas por una propiedad de evolución del siguiente estilo:
 $$x(t;x_0,t_0) = x(t;x(s;x_0,t_0),s)  $$
 para toda $t_0 \leq s \leq t$, lo cuál indica que este futuro está determinado completamente por el presente. Con el pasado metido sólo en determinar el presente. Esta es una versión determinista de la propiedad de Markov.

 Después de la explicación de Einstein de la observación del movimiento Browniano durante la primera década del siglo XX, hbieron varios intentos por otros físicos para formular la dinámica de tal movimiento en términos de ecuaciones diferenciales. Las ecuaciones resultantes fueron escritas de la siguiente manera: $$ dX_t = a(t,X_t)dt + b(t,X_t)ξ_tdt $$

 <font color="" face="Comic Sans MS">

**Procesos Simples**

 Análogamente a comenzar a definir una integral de Lebesgue, primero para funciones simples, definiremos la integral de Itô para procesos simples.

  **Definición** *Sea $0=t_0<t_1<...<t_n=T$ una partición finita del intervalo $[0,T]$. Un procesos estocástico simple es un proceso de la forma: $$X_t = \sum_{k=0}^{n-1}X^{(k)}1_{[t_k,t_{k+1}]}(t)  $$ en donde $X^{(i)}$ son una colección de variables aleatorias cuadrado integrables.*

  **Integral para Procesos Simples**
  
*Esta es la definición intuitiva de integral y establece que si el integrando es constante en algún subintervalo, entonces la integral debe ser esa constante multiplicada por el incremento del movimiento Browniano en dicho subintervalo.*

**Definición** *La integral estocástica de Itô de un proceso simple $X$ de la forma definida arriba, respecto del movimiento Browniano, denotada por* $$I(X) = \int_0^T X_s dB_s = \sum_{k=0}^{n-1} X^{(k)} (B_{t_{k+1}} - B_{t_{k}}) $$

  Observemos ahora que el movimiento Browniano es un ejemplo de un proceso medible y es posible demostrar que el proceso puede ser aproximado en el sentido de la norma del espacio $L^2(P\times dt)$ por el proceso simple:
 $$\sum_{k=0}^{n-1}B_{t_k} 1_{[t_k,t_{k+1}]}$$ Se tiene entonces la siguiente integral estocástica particular y su aproximación como límite en media cuadrática:
  $$\int_0^TB_tdB_t =\lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_k}(B_{t_{k+1}}-{B_{t_k}})  $$
  
  Aquí entendemos que el límite está en el sentido de la distancia máxima entre los dos puntos sucesivos de la partición tendiendo a cero. Al final llegamos a la versión más general de la integral. Simplemente haremos una pequeña extención con un pequeño artificio:
  $$I_t(X) = \int_0^TX_s1_{[0,t]}(s)dB_s = \int_0^tX_sdB_s$$ Ahora nos permitimos ver a la integral estocástica no como una variable aleatoria suma de variables aleatorias, si no ya como un proceso.

  Ahora, observemos al movimiento Browniano nuevamente, pues tiene sentido la expresión: $$\int_0^tf(B_s)dB_s $$ Y se puede demostrar que esta integral puede ser calculada mediante el sigiente límite en el espacio $L^2(P)$: $$\int_0^tf(B_s)dB_s =\lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_k}(B_{t_{k+1}}-{B_{t_k}})   $$
  
  Resolvamos por ejemplo, el caso particular: $f(B_s) = B_s$, es decir veamos qué ocurre con (1)
  Usando la identidad: $$ a(b-a) = \frac{1}{2}(b^2-a^2)-\frac{1}{2}(a-b)^2 $$
  \begin{eqnarray*}
  \int_0^tf(B_s)dB_s &=& \lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_k}(B_{t_{k+1}}-{B_{t_k}})\\
  &=& \sum_{k=0}^{n-1} \frac{1}{2}(B_{t_{k+1}}^2-B_{t_k}^2)-\frac{1}{2}(B_{t_k}-B_{t_{k+1}})^2\\
  &=& \frac{1}{2}B_t^2-\frac{1}{2}t
  \end{eqnarray*}
  Aparece un término usual conocido en una integral de Riemann, $\frac{1}{2}B_t^2$ Sin embargo, también aparece un término conocido como correción de Itô $-\frac{1}{2}t$. \\Si ahora tomamos los extremos por la derecha veamos qué sucede:
   \begin{eqnarray*}
  \int_0^tf(B_s)dB_s &=& \lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_{k+1}}(B_{t_{k+1}}-{B_{t_k}})\\
  &=& \sum_{k=0}^{n-1} \frac{1}{2}(B_{t_{k+1}}^2-B_{t_k}^2)+\frac{1}{2}(B_{t_k}-B_{t_{k+1}})^2\\
  &=& \frac{1}{2}B_t^2+\frac{1}{2}t
  \end{eqnarray*}
  
  Y llegamos así a un resultado curioso, la integral estocástica sí es sensible al punto en donde se evalúa el integrando, a diferencia de la integral de Riemann.
  
  Ahora uno podría preguntarse, ¿Qué sucede si ahora nos tomamos el punto medio?. Pues si lo hacemos de esta manera, entonces obtenemos la Integral de Stratonovich, que se denota así:$$\int_0^t B_s \circ dB_s = \frac{1}{2}B_t^2 $$ y desaparece así el término adicional de la integral de Itô. Es decir, tiene unas ventajas operacionales a diferencia de Itô, sin embago, ya no cumple con algunas otras propiedades importantes del cálculo estocástico, como la propíedad de ser una martingala.
  
  
  Otra propiedad importante de la integral de Itô es que es una función lineal:
  $$ \int_0^t (cX_s +Y_s) dB_s = c\int_0^tX_sdBs + \int_0^tY_sdB_s $$

```python
# Comenzamos a generar los valores que ocuparemos

# Número de divisiones en el intervalo
n = 100
#Pasos
pasos = 1
def aproxSumas(n: int, pasos: int):
  '''
  This function plots the random values from a normal distribution and plots
  the aproximation by Itô sums
  -n : int, this value is the number of discretization of the interval
  -pasos: int, this value is the number that goes from one interval to another
  '''
  np.random.seed(1)
  # Ahora vamos a generar los valores del movimiento Browniano, basándonos
  # en una normal y los vamos sumando conforme aparecen, como una suma
  # de los primeros n naturales
  bm = np.append([0], (np.random.normal(0,np.sqrt(pasos/n),n)).cumsum())

  # Generamos valores de referencia para los plots
  steps = np.linspace(0,pasos,n+1)

  # Definimos los valores que aproximan
  integral = (bm[0:n-1]*(bm[1:n]-bm[0:n-1]))
  integral1 = (bm[0:n-1]*(bm[1:n]-bm[0:n-1])).cumsum()
  plt.plot(steps[0:n-1], 0.5*(bm**2-steps)[0:n-1], label='Suma de Ito', color = 'black')
  plt.plot(steps[0:n-1], integral1, label='Aprox por áreas', color = 'purple')
  plt.plot(steps[0:n-1],integral, label='Variación')
  plt.xlabel('Pasos')
  plt.ylabel('Valor')
  plt.legend()
  plt.show()
```

```python
import math
import random
import matplotlib.pyplot as plt

N = 10000                      # numero de pasos
T0 = 0.0                       # limite izquierdo
T = 1.0                        # limite derecho
DELTA = (T - T0) / N           # incremento
X0 = 10.0                       #v valor inicial

def generate():
    u1 = random.uniform(0, 1)
    u2 = random.uniform(0, 1)
    r = math.sqrt(-2 * math.log(u1))
    theta = 2 * math.pi * u2
    return r * math.cos(theta)

sqdelta = math.sqrt(DELTA)
u2 = 0.0
wt = 0.0
rstk = 0.0
tk = T0
k = 0

xt = [0.0] * (N+1)
rst = [0.0] * (N+1)
abscissa = [0.0] * (N+1)

while k < N:
    k += 1
    if k % 2 == 1:
        u1, u2 = generate(), u2
    else:
        u1 = u2
    dwt = u1 * sqdelta                        # incremento de la trayectoria
    wt += dwt                                 # valor actual
    xt[k] = 0.5 * (wt * wt - tk)               # valor de la solución por fórmula
    rst[k] = rstk + (wt - dwt) * dwt           # valor de la suma de Itô
    tk += DELTA                               # tiempo
    abscissa[k] = tk                          # valor de los ejes
    rstk = rst[k]

plt.plot(abscissa, xt, label='Solución explícita')
plt.plot(abscissa, rst, label='Suma de Itô')
plt.xlabel('Tiempo')
plt.ylabel('Valor')
plt.legend()
plt.show()
```

## Ecuaciones

La integral de Itô nos provee con las herramientas necesarias para formular las ecuaciones estocásticas. Tales ecuaciones describen la dinámica de muchos sistemas estocásticos continuos muy importantes.
\\ Sea $(\Omega, \mathcal F, P)$ un espacio de proabilidad y sea $B_t$ un proceso de Wiener unidimensional:

\textbf{Definición.} Sea $b(t,X_t)$ y $\sigma(t,X_t)$ dos funciones de $[0,T] \times \mathbb R en \mathbb R$ Una ecuación estocástica es una ecuación de la forma $$dX_t = b(t,X_t)dt + \sigma(t,X_t)dB_t$$ definidas para los valores de t en el intervalo [0,T] y con condición inicial la variable aleatoria $X_0$ que se presupone independiente del movimiento Browniano. Esta ecuación se interpreta como una ecuación integral más que como una ecuación diferencial:
$$ X_t = X_0 + \int_0^tb(s,X_s)ds + \int_0^t\sigma(s,X_s)dB_s $$
En donde la primera es una integral de Riemann, mientras que la segunda es una integral estocástica de Itô. Al proceso $\{X_t\}$ se le llama proceso de Itô.

Los elementos conocidos de esta ecuación son los coeficientes $b, \sigma$, junto con la variable aleatoria inicial $X_0$. A la función $b(t,x)$ se le conoce como coeficiente de tendencia \textit{drift}, pues representa la tendencia hacia la que el fenómeno está yendo. Mientras que la función $\sigma(t,x)$ representa el coeficiente de difusión, en varia literatura también se le conoce como el ruido aditivo dado por la ecuación estocástica.

\textbf{Fórmula de Itô II}. sI $\{X_t\}$ es un proceso de Itô y $f(t,x)$ es una función de clase $C^1$ en t y de clase $C^2$ en x, entonces el proceso $\{Y_t = f(t,X_t)\}$ es también un proceso de Itô y satisface la ecuación estocástica:
$$ dY_t = f_t(t,X_t)dt + f_x(t,X_t)dX_t + \frac{1}{2}f_{xx}(dX_t)^2 $$ Los subíndices indican el orden de la derivada.

Después de realizar una serie de análisis en la fórmula de Itô, llegamos a la siguiente proposición:\\
Sea $U:[0,T]\times \mathbb R \rightarrow \mathbb R $ tiene derivadas parciales continuas, una transformación escalar de $U$ resulta en: $$ U(t,X_t) = U(s,X_s) + \int_s^t \{\frac{\partial U}{\partial t} + e\frac{\partial U}{\partial x} + \frac{1}{2}f^2 \frac{\partial^2U}{\partial X^2}\}du +\int_s^tf\frac{\partial U}{\partial x} dW_u \quad \quad \quad (a)$$

Utilizando este resultado y haciendo uso de ecuaciones parciales, llegamos a lo siguiente:


Si $$ X_t = X_0 + \int_0^tb(s,X_s)ds + \int_0^t\sigma(s,X_s)dB_s $$ es un proceso de Itô con ruido multiplicativo y es una ecuación lineal, usando la identidad de Itô (a)m entonces tiene una solución explícita de la siguiente manera:
\label{sec:resultado1}
$$\label{sec:resultado1} X_t = X_0exp(a-\frac{1}{2}b^2)t+bB_t  \quad \quad \quad \quad (b)$$

```python
# Importamos las bibliotecas necesarias para el botón deslizante
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

#Definimos los parámetros de la ecuación

#Puntos final e inicial del intervalo
T0 = 0
T = 1
#Punto inicial
X0 = 1
#coeficientes de la ecuación
A = 1
B = A
WPTO = 0

def explsol(tt, wpt):
    return X0 * np.exp((A - 0.5 * B * B) * (tt - T0) + B * (wpt - WPTO))

def plot_solution(numinv):
    DELTA = (T - T0) / numinv
    np.random.seed(1)
    abscissa = np.linspace(T0, T, num=numinv+1)
    xt = np.zeros(numinv+1)

    sqdelta = np.sqrt(DELTA)
    u2 = np.random.standard_normal()
    wt = WPTO
    tk = T0
    k = 0

    while k <= numinv:
        tk += DELTA
        abscissa[k] = tk
        u1 = np.random.standard_normal()
        wt += u1 * sqdelta
        xt[k] = explsol(tk, wt)
        u2 = u1
        k += 1

    plt.plot(abscissa, xt)
    plt.xlabel('t')
    plt.ylabel('X(t)')
    plt.title('Solución explícita de una SDE lineal')
    plt.show()

numinv_slider = widgets.IntSlider(min=100, max=1000, step=50, value=500, description='NUMINV:')
widgets.interact(plot_solution, numinv=numinv_slider)
```

Y de aquí utilizamos el Movimiento Browniano Geométrico. Este modelo es de amplio uso en finanzas y sirve para representar el precio de algunos bienes que fluctúan siguiendo los vaivenes de los mercados financieros. Se define de la siguiente manera:

Sean $\mu y \sigma>0$ dos constantes, y $x_0>0$. El movimiento Browniano geométrico es el proceso de la ecuación estocástica:$$ dX_t = \mu X_tdt + \sigma X_tdB_t $$$$X_0 = x_0$$
y por tanto, tiene solución:$$ X_t = X_0exp(\mu-\frac{1}{2}\sigma^2)t+bB_t  $$

Nótese que en ausencia del ruido, se reduce a la ecuación conocida, la cuál tiene solución $X_t = x_0e^{\mu t}$ Esta función representa el comportamiento en el tiempo un capital inicial positivo que crece de forma determinista.

## ***Metodología***

Comenzamos con la aproximación de Euler-Maruyama.

Euler-Maruyama nEs un método de discretización para obtener las trayectorias de una ecuación estocástica de la forma: $$dX_t = b(t,X_t)dt+\sigma(t,X_t)dB_t \quad \quad \quad (2)$$ $$X_0 = x_0$$
En este procedimiento se divide el intervalo [0,t] de manera uniforma en n subintervalos de indéntica longitud $\Delta t = \frac{t}{n}$ y se define $t_j = j\Delta_t$ para $j = 0,1,2,...,N$. Suponiendo que $Y_j$ es un valor al azar de la distribución normal estandar, se definen los valores sucesivos de la trayectoria solución de la ecuación estocástica como sigue:
$$X_0 = x_0 $$ $$ X_{t_{j+1}} = X_{t_j} + b(t_j,Xt_j) \Delta t + \sigma (t_j, X_{t_j}) \sqrt{\Delta t }Y_j $$

Como puede observarse, es una extensión del método común de Euler. Esto siempre que la función $\sigma := 0$, entonces ahí ya tenemos una ecuación determinista.\\
Aquí nosotros calclaremos nuestra propia trayectoria Browniana discretizada y la usaremos para calcular los incrementos que se requieren en la fórmula, pues terminaríamos calculándolo de esta manera:
$$X_J = X_{j-1} + f(X_{j-1})\Delta t + \sigma(X_{j-1})(W(t_j)-W(t_{j-1})) $$

```python
# Aquí vamos a comparar los resultados entre una ecuación lineal estocástica
# y su aproximación por el método de discretización de Euler-M

np.random.seed(1000)
lambda_val = 2
mu = 1
Xzero = 1

T = 1
N = 1000
dt = 1/N

dW = np.sqrt(dt) * np.random.randn(N)
W = np.cumsum(dW)

Xtrue = Xzero * np.exp((lambda_val - 0.5 * mu**2) * np.arange(dt, T + dt, dt) + mu * W)

plt.plot(np.arange(0, T + dt, dt), np.concatenate(([Xzero], Xtrue)), 'm-', label='Verdadera solución')

R = 4
Dt = R * dt
L = N // R

Xem = np.zeros(L)
Xtemp = Xzero

for j in range(L):
    Winc = np.sum(dW[R*(j-1)+1:R*j+1])
    Xtemp = Xtemp + Dt * lambda_val * Xtemp + mu * Xtemp * Winc
    Xem[j] = Xtemp

plt.plot(np.arange(0, T + Dt, Dt), np.concatenate(([Xzero], Xem)),  label='Aproximación EM')

plt.xlabel('t', fontsize=12)
plt.ylabel('X', fontsize=16, rotation=0, ha='right')
plt.legend()
plt.show()
```


Por último, veremos la justificación teórica de un proceso muy importante dentro de las ecuaciones.

Consideremos un un caso lineal simple, en el que bombardeamos un grano de polvo sobre una superficie de agua, esto responde al movimiento Browniano. La intensidad de este bombardeo no depende de las variables estáticas, por ejemplo, la posición, la velocidad o el grano. Tomando $X_t$ como uno de los componentes de la partícula, Paul Langevin escribió la ecuación: $$ \frac{dX_t}{dt} = -aX_t + + b\xi_t  $$
Ahora interpretaremos la ecuación como:
$$ X_t = X_0 - \int_0^taX_sds + \int_0^tbdW_s $$


Y nuevamente, usando la formula, esto tiene solución $$X_t= X_0exp((a-\frac{1}{2}b^2)t+bW_t)  $$

```python
# Esta es una simulación adaptada para realizar el proceso Ornstein_Uhlenbeck

def ornstein_uhlenbeck(x0, theta, mu, sigma, dt, num_steps):
    x = np.zeros(num_steps)
    x[0] = x0
    for t in range(1, num_steps):
        drift = theta * (mu - x[t-1]) * dt
        diffusion = sigma * np.sqrt(dt) * np.random.normal()
        x[t] = x[t-1] + drift + diffusion
    return x

# Parámetros del proceso de Ornstein-Uhlenbeck
x0 = 0
theta = 0.5
mu = 1
sigma = 0.2
dt = 0.01
num_steps = 1000

# Simulación del proceso de Ornstein-Uhlenbeck
x = ornstein_uhlenbeck(x0, theta, mu, sigma, dt, num_steps)

# Visualización de la simulación
t = np.arange(0, num_steps) * dt
plt.plot(t, x)
plt.xlabel('Tiempo')
plt.ylabel('Valor del proceso')
plt.title('Simulación del proceso de Ornstein-Uhlenbeck')
plt.show()
```

Expansión de Ito-Taylor}
Recordemos primero la expansión de Taylor con la que somos familiares:

$$f[X(t(] = f[X(t_0)]+ \mathcal L f[X(t_0)](t-t_0)+\frac{1}{2}\mathcal L^2 f[X(t_0)](t-t_0)^2 + O(\mathcal L^3)  $$
Donde definimos los operadores: $$\mathcal L := a(X)\frac{\partial}{\partial X} $$
Ahora veamos una expresión similar, pero para una ecuación estocástica:

$$dX(t) = a[X(t)]dt+b[X(t)]dW(t) $$

De igual manera que con los ejemplos, por simplicidad se considerará el caso: $a= a[X(t)]. b= b[X(t)]$ Y entonces la fórmula de Itô nos lleva a lo siguiente:

\begin{equation}
    df[X(t)] = \{a \frac{\partial}{\partial X}F[X(t)] + \frac{1}{2}b^2[X(t)]\frac{\partial^2}{\partial X^2}f[X(t)]\}dt + b[X(t)]\frac{\partial}{\partial X}f[X(t)]dW_t
\end{equation}

Y ahora defiminos:
$$\mathcal L^0 := a\frac{\partial}{\partial X}+ \frac{1}{2}b^2[X]\frac{\partial^2}{\partial X^2} $$
$$\mathcal L^1 := b[X]\frac{\partial }{\partial X}$$

Entonces (1) pasa a ser:
$$ df[X(t)] = \mathcal{L}^0 f[X(t)]dt + \mathcal L^1 f[X(t)]dW(t)$$
En su forma integral:

$$ f[X(t)] =f[X(t_0)] + \int_{t_0}^t \mathcal{L}^0 f[X(s)]ds +\int_{t_0}^t \mathcal L^1 f[X(s)]dW(s) \quad \quad \quad (2) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la ecuación estocástica
X0 = 1.0
T = 1.0
N = 1000
dt = T / N

# Generar una secuencia de números aleatorios para dW
np.random.seed(0)
dW = np.sqrt(dt) * np.random.randn(N)

# Inicializar el vector X para ambos esquemas
X_itotaylor = np.zeros(N)
X_euler = np.zeros(N)
X_itotaylor[0] = X0
X_euler[0] = X0

# Resolver la ecuación estocástica utilizando los esquemas de Ito-Taylor y Euler-Maruyama
for i in range(1, N):
    dX_itotaylor = -X_itotaylor[i-1]**3 * dt + np.sqrt(dt) * dW[i-1]
    dX_euler = -X_euler[i-1]**3 * dt + np.sqrt(dt) * dW[i-1]
    X_itotaylor[i] = X_itotaylor[i-1] + dX_itotaylor
    X_euler[i] = X_euler[i-1] + dX_euler

# Graficar los resultados
t = np.linspace(0, T, N)
plt.plot(t, X_itotaylor, label='Ito-Taylor')
plt.plot(t, X_euler, label='Euler-Maruyama')
plt.xlabel('Tiempo')
plt.ylabel('X')
plt.title('Solución de la ecuación estocástica')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_maruyama_simulation(S0, sigma0, mu, r, T, N, M):
    Delta = T / N
    sqrtdt = np.sqrt(Delta)

    Xem1 = np.ones((M, 1)) * S0
    Xem2 = np.ones((M, 1)) * sigma0

    for j in range(N):
        Winc1 = sqrtdt * np.random.randn(M, 1)
        Winc2 = sqrtdt * np.random.randn(M, 1)

        Xem1 = np.abs(Xem1 + Delta * mu * Xem1 + np.sqrt(Xem1) * Xem2 * Winc1)
        Xem2 = np.abs(Xem2 + Delta * 0.5 * (sigma0 - Xem2) + 2 * np.sqrt(Xem2) * Winc2)

    return Xem1, Xem2

def european_option_price(S, r, T, option_type='call'):
    if option_type == 'call':
        payoff = np.maximum(S - 1, 0)
    else:
        payoff = np.maximum(1 - S, 0)

    discounted_payoff = np.exp(-r * T) * payoff
    option_price = np.mean(discounted_payoff)

    return option_price

# Parameters
S0 = 100
sigma0 = 0.3
mu = 0.16
r = 0.05
T = 1.0
N = 2**10
M = int(1e4)

# Euler-Maruyama simulation
Xem1, Xem2 = euler_maruyama_simulation(S0, sigma0, mu, r, T, N, M)

# European Call Option Price
call_option_price = european_option_price(Xem1, r, T, option_type='call')
print("European Call Option Price:", call_option_price)

# European Put Option Price
put_option_price = european_option_price(Xem1, r, T, option_type='put')
print("European Put Option Price:", put_option_price)
```
