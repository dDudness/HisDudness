<font color="" face="Times New Roman" size="5px">**Definicin 1.** *Un proceso estocstico es una coleccin de variables aleatorias $\{X_t: t \in T\}$ parametrizada por un conjunto $T$, llamado espacio parametral, en donde las variables toman valores en un conjunto $S$ llamado espacio de estados.*

<font color="" face="Comic Sans MS">

Considrese un sistema que puede cambiar de un estado a otro, a travs del tiempo gracias a una ley de movimiento. Tambin considere que el sistema evoluciona de una manera no determinista, sino, azarosa, entonces obtenemos una coleccin de variables aleatorias que dependen de un parmetro $t$ que nos ayuda a indicar el estado $s$ en donde se encuentra nuestra variable.

La definicin de proceso estocstico toma como base un espacio de probabilidad $(\Omega, \mathscr{F}, \mathbb P) $ Y es importante tenerlo en cuenta, para enunciar de una manera ms especfica los resultados que se presentan ms adelante.

Particularmente, un proceso estocstico puede clasificarse segn dos tipos: Procesos a tiempo discreto y Procesos a tiempo continuo. Es importante sealar la diferencia entre el espacio de estados de una variable aleatoria y el espacio parametral. El espacio de estados tambin se puede pensar como un soporte de nuestra sucesin de variables parametrizada por el espacio $T$. Es decir, el proceso puede tener un soporte finito o no. En cambio, el espacio parametral discreto, tiene sus valores en los naturales o en los enteros $\mathbb{( N, Z)}$; en cambio, el espacio continuo generalmente se indica slo bajo el conjunto: $\{t\in T| t\geq0\}$.

<font color="" face="Comic Sans MS">

**Procesos de Markov**

Estos tipos de procesos son modelos de mucha importancia y son los de mayor estudio en la rama de la probabilidad. Suponiendo que se conoce el estado presente del sistema, entonces los estados anteriores no tienen influencia en los estados futuros del sistema. Esta condicin se llama propiedad de Markov y puede expresarse de la siguiente forma: Para cualesquiera tiempos $0\leq s\leq t$ y sea $A\in B(\mathbb R)$:  

$$P( X_t \in A| \mathcal F_s) = P(X_t \in A | X_s) $$

<font color="" face="Comic Sans MS">

Para poder hablar de variables aleatorias con valores en el espacio de estados $S$, es necesario asociar a este conjunto una $\sigma$-lgebra.
Considerando que $S$ es un subconjunto de $\mathbb R$, puede tomarse a la $\sigma$-lgebra de Borel de R restringida a S, es decir, $S \cap \mathscr B(\mathbb R)$.

Un proceso estocstico puede considerarse como una funcin de dos variables: $$ X: T  x  \Omega  S $$
As, al par ordenado $(t,\omega)$ se le asocia el valor $X(t,\omega)$, simplificando la notacin se escribe: $ X_t(\omega) $. Para cada valor de $t$ en $T$, el mapeo $\omega  X_t(\omega)$ es una v.a. Mientras que para cada $\omega$ en $\Omega$ fijo, la funcin $t \rightarrow X_t(\omega)$ es llamada una trayectoria o realizacin del proceso. Es por ello que a veces se define un proceso estocstico como una funcin aleatoria. Una de tales trayectorias tpicas que, adems cuenta con la propiedad de ser continua se muestra en la figura siguiente y corresponde a una trayectoria de un Movimiento Browniano.

<font color="" face="Comic Sans MS">

**Movimiento Browniano**

**Definicin 3.** Un movimiento Browniano estndar unidimensional es un proceso estocstico $\{B_t: t\geq 0\}$ tal que:


1.   $B_0 = 0$ casi seguramente.
2.   Las trayectorias $t  B_t$ son continuas.
3.   El proceso tiene incrementos independientes.
4.   La variable $B_t - B_s$ tiene distribucin $N(0,t-s)$ para $0\leq s\leq t$.

Las condiciones que aparecen en esta definicin fueron dadas con consecuencia directa de las observaciones del fenmeno fsico que explic Einstein. Sin embargo, no garantiza que tal objeto matemtico exista. En 1923 el matemtico norteamericano Norbert Wiener demostr la existencia de un proceso con tales condiciones. Por lo que tambin es llamado as, Proceso De Wiener.

Ahora veremos algunos resultados importantes de este proceso.

Funcin de Probabilidad de transicin:

En particular, la probabilidad de que un movimiento Browniano que inicia en x se encuentre en un conjunto $A \subseteq \mathbb R$ despus de t unidades de tiempo es: $$ p(t,x,A) = \int_A p(t,x,y)dy  $$

Donde tambin definimos: $$p(t,x,y) = \frac{1}{\sqrt{2\pi\sigma ^2t }} e^{-(y-x)^2 / 2 \sigma ^2t}$$


De la propiedad 4, podemos hacer $s=0$ y entonces obtenemos que en particular la variable $B_t \backsim N(0,t)$. y por tanto:
1.  $E(B_t) = 0$
2.  $Var(B_t) = E(B_t^2) = t$
3.  $Var(B_t-B_s) = E((B_t-B_s)^2) = t-s$

Tambin es importante resaltar que el Proceso de Wiener cumple con la propiedad de Markov.

**Procesos Simples**

  **Definicin** *Sea $0=t_0<t_1<...<t_n=T$ una particin finita del intervalo $[0,T]$. Un procesos estocstico simple es un proceso de la forma: $$X_t = \sum_{k=0}^{n-1}X^{(k)}1_{[t_k,t_{k+1}]}(t)  $$ en donde $X^{(i)}$ son una coleccin de variables aleatorias cuadrado integrables.*

  **Integral para Procesos Simples**
  
*Esta es la definicin intuitiva de integral y establece que si el integrando es constante en algn subintervalo, entonces la integral debe ser esa constante multiplicada por el incremento del movimiento Browniano en dicho subintervalo.*

**Definicin** *La integral estocstica de It de un proceso simple $X$ de la forma definida arriba, respecto del movimiento Browniano, denotada por* $$I(X) = \int_0^T X_s dB_s = \sum_{k=0}^{n-1} X^{(k)} (B_{t_{k+1}} - B_{t_{k}}) $$

  Observemos ahora que el movimiento Browniano es un ejemplo de un proceso medible y es posible demostrar que el proceso puede ser aproximado en el sentido de la norma del espacio $L^2(P\times dt)$ por el proceso simple:
 $$\sum_{k=0}^{n-1}B_{t_k} 1_{[t_k,t_{k+1}]}$$ Se tiene entonces la siguiente integral estocstica particular y su aproximacin como lmite en media cuadrtica:
  $$\int_0^TB_tdB_t =\lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_k}(B_{t_{k+1}}-{B_{t_k}})  $$
  
  Aqu entendemos que el lmite est en el sentido de la distancia mxima entre los dos puntos sucesivos de la particin tendiendo a cero. Al final llegamos a la versin ms general de la integral. Simplemente haremos una pequea extencin con un pequeo artificio:
  $$I_t(X) = \int_0^TX_s1_{[0,t]}(s)dB_s = \int_0^tX_sdB_s$$ Ahora nos permitimos ver a la integral estocstica no como una variable aleatoria suma de variables aleatorias, si no ya como un proceso.

  Ahora, observemos al movimiento Browniano nuevamente, pues tiene sentido la expresin: $$\int_0^tf(B_s)dB_s $$ Y se puede demostrar que esta integral puede ser calculada mediante el sigiente lmite en el espacio $L^2(P)$: $$\int_0^tf(B_s)dB_s =\lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_k}(B_{t_{k+1}}-{B_{t_k}})   $$
  
  Resolvamos por ejemplo, el caso particular: $f(B_s) = B_s$, es decir veamos qu ocurre con (1)
  Usando la identidad: $$ a(b-a) = \frac{1}{2}(b^2-a^2)-\frac{1}{2}(a-b)^2 $$
  \begin{eqnarray*}
  \int_0^tf(B_s)dB_s &=& \lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_k}(B_{t_{k+1}}-{B_{t_k}})\\
  &=& \sum_{k=0}^{n-1} \frac{1}{2}(B_{t_{k+1}}^2-B_{t_k}^2)-\frac{1}{2}(B_{t_k}-B_{t_{k+1}})^2\\
  &=& \frac{1}{2}B_t^2-\frac{1}{2}t
  \end{eqnarray*}
  Aparece un trmino usual conocido en una integral de Riemann, $\frac{1}{2}B_t^2$ Sin embargo, tambin aparece un trmino conocido como correcin de It $-\frac{1}{2}t$. \\Si ahora tomamos los extremos por la derecha veamos qu sucede:
   \begin{eqnarray*}
  \int_0^tf(B_s)dB_s &=& \lim_{n\rightarrow \infty} \sum_{k=0}^{n-1}B_{t_{k+1}}(B_{t_{k+1}}-{B_{t_k}})\\
  &=& \sum_{k=0}^{n-1} \frac{1}{2}(B_{t_{k+1}}^2-B_{t_k}^2)+\frac{1}{2}(B_{t_k}-B_{t_{k+1}})^2\\
  &=& \frac{1}{2}B_t^2+\frac{1}{2}t
  \end{eqnarray*}
  
  Y llegamos as a un resultado curioso, la integral estocstica s es sensible al punto en donde se evala el integrando, a diferencia de la integral de Riemann.
  
  Ahora uno podra preguntarse, Qu sucede si ahora nos tomamos el punto medio?. Pues si lo hacemos de esta manera, entonces obtenemos la Integral de Stratonovich, que se denota as:$$\int_0^t B_s \circ dB_s = \frac{1}{2}B_t^2 $$ y desaparece as el trmino adicional de la integral de It. Es decir, tiene unas ventajas operacionales a diferencia de It, sin embago, ya no cumple con algunas otras propiedades importantes del clculo estocstico, como la propedad de ser una martingala.

## Ecuaciones

La integral de It nos provee con las herramientas necesarias para formular las ecuaciones estocsticas. Tales ecuaciones describen la dinmica de muchos sistemas estocsticos continuos muy importantes.
\\ Sea $(\Omega, \mathcal F, P)$ un espacio de proabilidad y sea $B_t$ un proceso de Wiener unidimensional:

\textbf{Definicin.} Sea $b(t,X_t)$ y $\sigma(t,X_t)$ dos funciones de $[0,T] \times \mathbb R en \mathbb R$ Una ecuacin estocstica es una ecuacin de la forma $$dX_t = b(t,X_t)dt + \sigma(t,X_t)dB_t$$ definidas para los valores de t en el intervalo [0,T] y con condicin inicial la variable aleatoria $X_0$ que se presupone independiente del movimiento Browniano. Esta ecuacin se interpreta como una ecuacin integral ms que como una ecuacin diferencial:
$$ X_t = X_0 + \int_0^tb(s,X_s)ds + \int_0^t\sigma(s,X_s)dB_s $$
En donde la primera es una integral de Riemann, mientras que la segunda es una integral estocstica de It. Al proceso $\{X_t\}$ se le llama proceso de It.

Los elementos conocidos de esta ecuacin son los coeficientes $b, \sigma$, junto con la variable aleatoria inicial $X_0$. A la funcin $b(t,x)$ se le conoce como coeficiente de tendencia \textit{drift}, pues representa la tendencia hacia la que el fenmeno est yendo. Mientras que la funcin $\sigma(t,x)$ representa el coeficiente de difusin, en varia literatura tambin se le conoce como el ruido aditivo dado por la ecuacin estocstica.

\textbf{Frmula de It II}. sI $\{X_t\}$ es un proceso de It y $f(t,x)$ es una funcin de clase $C^1$ en t y de clase $C^2$ en x, entonces el proceso $\{Y_t = f(t,X_t)\}$ es tambin un proceso de It y satisface la ecuacin estocstica:
$$ dY_t = f_t(t,X_t)dt + f_x(t,X_t)dX_t + \frac{1}{2}f_{xx}(dX_t)^2 $$ Los subndices indican el orden de la derivada.

Despus de realizar una serie de anlisis en la frmula de It, llegamos a la siguiente proposicin:\\
Sea $U:[0,T]\times \mathbb R \rightarrow \mathbb R $ tiene derivadas parciales continuas, una transformacin escalar de $U$ resulta en: $$ U(t,X_t) = U(s,X_s) + \int_s^t \{\frac{\partial U}{\partial t} + e\frac{\partial U}{\partial x} + \frac{1}{2}f^2 \frac{\partial^2U}{\partial X^2}\}du +\int_s^tf\frac{\partial U}{\partial x} dW_u \quad \quad \quad (a)$$

Utilizando este resultado y haciendo uso de ecuaciones parciales, llegamos a lo siguiente:


Si $$ X_t = X_0 + \int_0^tb(s,X_s)ds + \int_0^t\sigma(s,X_s)dB_s $$ es un proceso de It con ruido multiplicativo y es una ecuacin lineal, usando la identidad de It (a)m entonces tiene una solucin explcita de la siguiente manera:
\label{sec:resultado1}
$$\label{sec:resultado1} X_t = X_0exp(a-\frac{1}{2}b^2)t+bB_t  \quad \quad \quad \quad (b)$$

```python
# Importamos las bibliotecas necesarias para el botn deslizante
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from IPython.display import display

#Definimos los parmetros de la ecuacin

#Puntos final e inicial del intervalo
T0 = 0
T = 1
#Punto inicial
X0 = 1
#coeficientes de la ecuacin
A = 1
B = A
WPTO = 0

def explsol(tt, wpt):
    return X0 * np.exp((A - 0.5 * B * B) * (tt - T0) + B * (wpt - WPTO))

def plot_solution(numinv):
    DELTA = (T - T0) / numinv
    np.random.seed(1)
    abscissa = np.linspace(T0, T, num=numinv+1)
    xt = np.zeros(numinv+1)

    sqdelta = np.sqrt(DELTA)
    u2 = np.random.standard_normal()
    wt = WPTO
    tk = T0
    k = 0

    while k <= numinv:
        tk += DELTA
        abscissa[k] = tk
        u1 = np.random.standard_normal()
        wt += u1 * sqdelta
        xt[k] = explsol(tk, wt)
        u2 = u1
        k += 1

    plt.plot(abscissa, xt)
    plt.xlabel('t')
    plt.ylabel('X(t)')
    plt.title('Solucin explcita de una SDE lineal')
    plt.show()

numinv_slider = widgets.IntSlider(min=100, max=1000, step=50, value=500, description='NUMINV:')
widgets.interact(plot_solution, numinv=numinv_slider)
```

Y de aqu utilizamos el Movimiento Browniano Geomtrico. Este modelo es de amplio uso en finanzas y sirve para representar el precio de algunos bienes que fluctan siguiendo los vaivenes de los mercados financieros. Se define de la siguiente manera:

Sean $\mu y \sigma>0$ dos constantes, y $x_0>0$. El movimiento Browniano geomtrico es el proceso de la ecuacin estocstica:$$ dX_t = \mu X_tdt + \sigma X_tdB_t $$$$X_0 = x_0$$
y por tanto, tiene solucin:$$ X_t = X_0exp(\mu-\frac{1}{2}\sigma^2)t+bB_t  $$

Ntese que en ausencia del ruido, se reduce a la ecuacin conocida, la cul tiene solucin $X_t = x_0e^{\mu t}$ Esta funcin representa el comportamiento en el tiempo un capital inicial positivo que crece de forma determinista.

```python
# En esta grfica estamos realizando la simulacin sola del movimiento browniano geomtrico.

#np.random.seed(100)
T = 1
N = 300
dt = T / N
xcero = 1
mu = 1
sigma =1/3
dW = np.zeros(N)
MBG = np.zeros(N)
dW[0] = np.sqrt(dt) * np.random.randn()
MBG[0] = xcero + mu * dt + sigma * xcero * dW[0]

for j in range(1, N):
    dW[j] = np.sqrt(dt) * np.random.randn()
    MBG[j] = MBG[j-1] + mu * MBG[j-1] * dt + sigma * MBG[j-1] * dW[j]

t = np.arange(0, T + dt, dt)
plt.plot(t, np.concatenate(([xcero], MBG)), 'r-')
plt.xlabel('Time')
plt.ylabel('MBG')
plt.show()
```

## ***Metodologa***

Comenzamos con la aproximacin de Euler-Maruyama.

Euler-Maruyama nEs un mtodo de discretizacin para obtener las trayectorias de una ecuacin estocstica de la forma: $$dX_t = b(t,X_t)dt+\sigma(t,X_t)dB_t \quad \quad \quad (2)$$ $$X_0 = x_0$$
En este procedimiento se divide el intervalo [0,t] de manera uniforma en n subintervalos de indntica longitud $\Delta t = \frac{t}{n}$ y se define $t_j = j\Delta_t$ para $j = 0,1,2,...,N$. Suponiendo que $Y_j$ es un valor al azar de la distribucin normal estandar, se definen los valores sucesivos de la trayectoria solucin de la ecuacin estocstica como sigue:
$$X_0 = x_0 $$ $$ X_{t_{j+1}} = X_{t_j} + b(t_j,Xt_j) \Delta t + \sigma (t_j, X_{t_j}) \sqrt{\Delta t }Y_j $$

Como puede observarse, es una extensin del mtodo comn de Euler. Esto siempre que la funcin $\sigma := 0$, entonces ah ya tenemos una ecuacin determinista.\\
Aqu nosotros calclaremos nuestra propia trayectoria Browniana discretizada y la usaremos para calcular los incrementos que se requieren en la frmula, pues terminaramos calculndolo de esta manera:
$$X_J = X_{j-1} + f(X_{j-1})\Delta t + \sigma(X_{j-1})(W(t_j)-W(t_{j-1})) $$

```python
# Aqu vamos a comparar los resultados entre una ecuacin lineal estocstica
# y su aproximacin por el mtodo de discretizacin de Euler-M

np.random.seed(1000)
lambda_val = 2
mu = 1
Xzero = 1

T = 1
N = 1000
dt = 1/N

dW = np.sqrt(dt) * np.random.randn(N)
W = np.cumsum(dW)

Xtrue = Xzero * np.exp((lambda_val - 0.5 * mu**2) * np.arange(dt, T + dt, dt) + mu * W)

plt.plot(np.arange(0, T + dt, dt), np.concatenate(([Xzero], Xtrue)), 'm-', label='Verdadera solucin')

R = 4
Dt = R * dt
L = N // R

Xem = np.zeros(L)
Xtemp = Xzero

for j in range(L):
    Winc = np.sum(dW[R*(j-1)+1:R*j+1])
    Xtemp = Xtemp + Dt * lambda_val * Xtemp + mu * Xtemp * Winc
    Xem[j] = Xtemp

plt.plot(np.arange(0, T + Dt, Dt), np.concatenate(([Xzero], Xem)),  label='Aproximacin EM')

plt.xlabel('t', fontsize=12)
plt.ylabel('X', fontsize=16, rotation=0, ha='right')
plt.legend()
plt.show()
```


Por ltimo, veremos la justificacin terica de un proceso muy importante dentro de las ecuaciones.

Consideremos un un caso lineal simple, en el que bombardeamos un grano de polvo sobre una superficie de agua, esto responde al movimiento Browniano. La intensidad de este bombardeo no depende de las variables estticas, por ejemplo, la posicin, la velocidad o el grano. Tomando $X_t$ como uno de los componentes de la partcula, Paul Langevin escribi la ecuacin: $$ \frac{dX_t}{dt} = -aX_t + + b\xi_t  $$
Ahora interpretaremos la ecuacin como:
$$ X_t = X_0 - \int_0^taX_sds + \int_0^tbdW_s $$


Y nuevamente, usando la formula, esto tiene solucin $$X_t= X_0exp((a-\frac{1}{2}b^2)t+bW_t)  $$

```python
# Esta es una simulacin adaptada para realizar el proceso Ornstein_Uhlenbeck

def ornstein_uhlenbeck(x0, theta, mu, sigma, dt, num_steps):
    x = np.zeros(num_steps)
    x[0] = x0
    for t in range(1, num_steps):
        drift = theta * (mu - x[t-1]) * dt
        diffusion = sigma * np.sqrt(dt) * np.random.normal()
        x[t] = x[t-1] + drift + diffusion
    return x

# Parmetros del proceso de Ornstein-Uhlenbeck
x0 = 0
theta = 0.5
mu = 1
sigma = 0.2
dt = 0.01
num_steps = 1000

# Simulacin del proceso de Ornstein-Uhlenbeck
x = ornstein_uhlenbeck(x0, theta, mu, sigma, dt, num_steps)

# Visualizacin de la simulacin
t = np.arange(0, num_steps) * dt
plt.plot(t, x)
plt.xlabel('Tiempo')
plt.ylabel('Valor del proceso')
plt.title('Simulacin del proceso de Ornstein-Uhlenbeck')
plt.show()
```

Expansin de Ito-Taylor}
Recordemos primero la expansin de Taylor con la que somos familiares:

$$f[X(t(] = f[X(t_0)]+ \mathcal L f[X(t_0)](t-t_0)+\frac{1}{2}\mathcal L^2 f[X(t_0)](t-t_0)^2 + O(\mathcal L^3)  $$
Donde definimos los operadores: $$\mathcal L := a(X)\frac{\partial}{\partial X} $$
Ahora veamos una expresin similar, pero para una ecuacin estocstica:

$$dX(t) = a[X(t)]dt+b[X(t)]dW(t) $$

De igual manera que con los ejemplos, por simplicidad se considerar el caso: $a= a[X(t)]. b= b[X(t)]$ Y entonces la frmula de It nos lleva a lo siguiente:

\begin{equation}
    df[X(t)] = \{a \frac{\partial}{\partial X}F[X(t)] + \frac{1}{2}b^2[X(t)]\frac{\partial^2}{\partial X^2}f[X(t)]\}dt + b[X(t)]\frac{\partial}{\partial X}f[X(t)]dW_t
\end{equation}

Y ahora defiminos:
$$\mathcal L^0 := a\frac{\partial}{\partial X}+ \frac{1}{2}b^2[X]\frac{\partial^2}{\partial X^2} $$
$$\mathcal L^1 := b[X]\frac{\partial }{\partial X}$$

Entonces (1) pasa a ser:
$$ df[X(t)] = \mathcal{L}^0 f[X(t)]dt + \mathcal L^1 f[X(t)]dW(t)$$
En su forma integral:

$$ f[X(t)] =f[X(t_0)] + \int_{t_0}^t \mathcal{L}^0 f[X(s)]ds +\int_{t_0}^t \mathcal L^1 f[X(s)]dW(s) \quad \quad \quad (2) $$

```python
import numpy as np
import matplotlib.pyplot as plt

# Parmetros de la ecuacin estocstica
X0 = 1.0
T = 1.0
N = 1000
dt = T / N

# Generar una secuencia de nmeros aleatorios para dW
np.random.seed(0)
dW = np.sqrt(dt) * np.random.randn(N)

# Inicializar el vector X para ambos esquemas
X_itotaylor = np.zeros(N)
X_euler = np.zeros(N)
X_itotaylor[0] = X0
X_euler[0] = X0

# Resolver la ecuacin estocstica utilizando los esquemas de Ito-Taylor y Euler-Maruyama
for i in range(1, N):
    dX_itotaylor = -X_itotaylor[i-1]**3 * dt + np.sqrt(dt) * dW[i-1]
    dX_euler = -X_euler[i-1]**3 * dt + np.sqrt(dt) * dW[i-1]
    X_itotaylor[i] = X_itotaylor[i-1] + dX_itotaylor
    X_euler[i] = X_euler[i-1] + dX_euler

# Graficar los resultados
t = np.linspace(0, T, N)
plt.plot(t, X_itotaylor, label='Ito-Taylor')
plt.plot(t, X_euler, label='Euler-Maruyama')
plt.xlabel('Tiempo')
plt.ylabel('X')
plt.title('Solucin de la ecuacin estocstica')
plt.legend()
plt.show()
```

```python
import numpy as np

# Euler-Maruyama method for stochastic volatility SDE asset price
# SDE is dS(1) = mu*S(1) dt + S(2)*sqrt(S(1)) dW(1), S(1)_0 = Szero(1)
# dS(2) = 1/2*(sigma_0-S(2)) dt + 2*sqrt(S(2)) dW(2), S(2)_0 = sigma_0
# S(1) is the asset price, S(2) represents the volatility
# European Option Price
# Vectorized across samples

np.random.seed(1)

T = 1
N = 1000
Delta = T / N
M = int(1000)
mu = 0.05
Szero = 100
sigma_zero = 0.2
r = 0.05

Xem1 = np.ones(M) * Szero
Xem2 = np.ones(M) * sigma_zero

for j in range(N):
    Winc1 = np.sqrt(Delta) * np.random.randn(M)
    Winc2 = np.sqrt(Delta) * np.random.randn(M)
    Xem1 = np.abs(Xem1 + Delta * mu * Xem1 + np.sqrt(Xem1) * Xem2 * Winc1)
    Xem2 = np.abs(Xem2 + Delta * 1/2 * (sigma_zero - Xem2) + 2 * np.sqrt(Xem2) * Winc2)

Price = np.exp(-r) * np.mean(np.maximum(0, 100-Xem1))

print("European Put Option Price:", Price)
```

```python
import numpy as np
import matplotlib.pyplot as plt

def euler_maruyama_simulation(S0, sigma0, mu, r, T, N, M):
    Delta = T / N
    sqrtdt = np.sqrt(Delta)

    Xem1 = np.ones((M, 1)) * S0
    Xem2 = np.ones((M, 1)) * sigma0

    for j in range(N):
        Winc1 = sqrtdt * np.random.randn(M, 1)
        Winc2 = sqrtdt * np.random.randn(M, 1)

        Xem1 = np.abs(Xem1 + Delta * mu * Xem1 + np.sqrt(Xem1) * Xem2 * Winc1)
        Xem2 = np.abs(Xem2 + Delta * 0.5 * (sigma0 - Xem2) + 2 * np.sqrt(Xem2) * Winc2)

    return Xem1, Xem2

def european_option_price(S, r, T, option_type='call'):
    if option_type == 'call':
        payoff = np.maximum(S - 1, 0)
    else:
        payoff = np.maximum(1 - S, 0)

    discounted_payoff = np.exp(-r * T) * payoff
    option_price = np.mean(discounted_payoff)

    return option_price

# Parameters
S0 = 100
sigma0 = 0.3
mu = 0.16
r = 0.05
T = 1.0
N = 2**10
M = int(1e4)

# Euler-Maruyama simulation
Xem1, Xem2 = euler_maruyama_simulation(S0, sigma0, mu, r, T, N, M)

# European Call Option Price
call_option_price = european_option_price(Xem1, r, T, option_type='call')
print("European Call Option Price:", call_option_price)

# European Put Option Price
put_option_price = european_option_price(Xem1, r, T, option_type='put')
print("European Put Option Price:", put_option_price)
```
